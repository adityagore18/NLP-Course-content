# Removing stop word

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
str="What it does: Regexes are used to search for specific character sequences, validate input, or perform 'find and replace' operation"
```

```python
def rmv_stopword(str):        
        new_text=[]
        for word in str.split():
            if word in stopwords.words('english'):
#                 print(word)
                new_text.append('')
            else:
                    new_text.append(word)
        x=new_text[:]
        
        new_text.clear()
        
        return " ".join(x)
```

```python
rmv_stopword("Formally, given a training sample of tweets and labels, where label '1' denotes the tweet is racist/sexist and label '0' denotes the tweet is not racist/sexist, your objective is to predict the labels on the test dataset.")
```

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_word(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

# Regular Expression(Regex)

1>A regular expression (regex) is a pattern that matches character combinations in strings. You can use regexes to check if a string matches a pattern, such as an email address or password. <br>
2>A regex is a sequence of sub-expressions that can match a portion or the entire input <br>
3>Regular expression literal:trough which we crete Regular Expression <br>
4>ou can use regexes with methods like exec(), test(), match(), matchAll(), replace(), replaceAll(), search(), and split(). <br>
1. Repeaters (  *, +, and { } )   <br>
These symbols act as repeaters and tell the computer that the preceding character is to be used for more than just one time. <br>

2. The asterisk symbol ( * ) <br>
It tells the computer to match the preceding character (or set of characters) for 0 or more times (upto infinite). <br>
 <br>
Example : The regular expression ab*c will give ac, abc, abbc, abbbc….and so on  <br>
 <br>
3. The Plus symbol ( + )  <br>
It tells the computer to repeat the preceding character (or set of characters) at atleast one or more times(up to infinite). <br>
 <br>
Example : The regular expression ab+c will give abc, abbc, <br>
abbbc, … and so on. <br>
4. The curly braces { … } <br> 
It tells the computer to repeat the preceding character (or set of characters) for as many times as the value inside this bracket. <br>
 <br>
Example : {2} means that the preceding character is to be repeated 2  <br>
times, {min,} means the preceding character is matches min or  more  <br>
times. {min,max} means that the preceding character is repeated at <br>
least min & at most max times. <br>
 <br>
5. Wildcard ( . )  <br>
The dot symbol can take the place of any other symbol, that is why it is called the wildcard character. <br>
 <br>
Example :  <br>
The Regular expression .* will tell the computer that any character <br>
can be used any number of times. <br>
 <br>
6. Optional character ( ? )  <br>
This symbol tells the computer that the preceding character may or may not be present in the string to be matched. <br>
 <br>
Example :  <br>
We may write the format for document file as – “docx?” <br>
The ‘?’ tells the computer that x may or may not be  <br>
present in the name of file format. <br>
 <br>
7. The caret ( ^ ) symbol ( Setting position for the match ) <br>
The caret symbol tells the computer that the match must start at the beginning of the string or line. <br>
 <br>
Example : ^\d{3} will match with patterns like "901" in "901-333-". <br>
 <br>
8.  The dollar ( $ ) symbol  <br>

It tells the computer that the match must occur at the end of the string or before \n at the end of the line or string. <br>
 <br>
Example : -\d{3}$  will match with patterns like "-333" in "-901-333". <br>
 <br>
9. Character Classes  <br>
 <br>
A character class matches any one of a set of characters. It is used to match the most basic element of a language like a letter, a digit, a space, a symbol, etc.  <br>
 <br>
 <br>
\s: matches any whitespace characters such as space and tab. <br>
\S: matches any non-whitespace characters. <br>
\d: matches any digit character. <br>
\D: matches any non-digit characters. <br>
\w : matches any word character (basically alpha-numeric) <br>
\W: matches any non-word character. <br>
\b: matches any word boundary (this would include spaces, dashes, commas, semi-colons, etc. <br>
[set_of_characters]: Matches any single character in set_of_characters. By default, the match is case-sensitive. <br>
 <br>
 <br>
Example : [abc] will match characters a,b and c in any string. <br>
 <br>
10. [^set_of_characters] Negation:  <br>
Matches any single character that is not in set_of_characters. By default, the match is case-sensitive. <br>
 <br>
Example : [^abc] will match any character except a,b,c .v <br>
 <br>
11. [first-last] Character range:  <br>
Matches any single character in the range from first to last. <br>
 <br>
Example : [a-zA-z] will match any character from a to z or A to Z. <br>
 <br>
12. The Escape Symbol (  \  )  <br>
If you want to match for the actual ‘+’, ‘.’ etc characters, add a backslash( \ ) before that character. This will tell the  <br>computer to treat the following character as a search character and consider it for a matching pattern. <br>
 <br>
Example : \d+[\+-x\*]\d+ will match patterns like "2+2" <br>
and "3*9" in "(2+2) * 3*9". <br>
 <br>
13. Grouping Characters ( )  <br>
A set of different symbols of a regular expression can be grouped together to act as a single unit and behave as a block, for  <br>this, you need to wrap the regular expression in the parenthesis( ). <br>
 <br>
Example : ([A-Z]\w+) contains two different elements of the regular  <br>
expression combined together. This expression will match any pattern  <br>
containing uppercase letter followed by any character. <br>
 <br>
14. Vertical Bar (  |  )  <br>
Matches any one element separated by the vertical bar (|) character. <br>
 <br> <br>
Example :  th(e|is|at) will match words - the, this and that. <br>
 <br>
15. \number  <br>
Backreference: allows a previously matched sub-expression(expression captured or enclosed within circular brackets ) to be  <br>identified subsequently in the same regular expression. \n means that group enclosed within the n-th bracket will be  <br>repeated at current position. <br>
 <br>
Example : ([a-z])\1 will match “ee” in Geek because the character  <br>
at second position is same as character at position 1 of the match. <br>
 <br>
16. Comment ( ?# comment )  <br>
Inline comment: The comment ends at the first closing parenthesis. <br>
 <br>
Example : \bA(?#This is an inline comment)\w+\b <br>
 <br>
17. # [to end of line]  <br>
X-mode comment. The comment starts at an unescaped # and continues to the end of the line. <br>
 <br>
Example :  (?x)\bA\w+\b#Matches words starting with A <br>

```python
def rmv_html_tag(text):
    pattern=re.compile('<.*?>')
    return pattern.sub('',text)
```

# Stop Word Removal

```python
from nltk.corpus import stopwords
```

```python
print(stopwords.words("english"))
```

```python
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words("english"))
def rmv_stop_word(text):
    stop_words = set(stopwords.words("english"))
    return " ".join([word for word in word_tokenize(text) if word not in stop_words])
```

```python
import time
start=time.time()
data['review'] = data['review'].apply(rmv_stop_word)
print(time.time()-start)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma=WordNetLemmatizer()
```

```python
def lemmatizer(text):
    str=""
    for word in text.split():
        str+=" "+lemma.lemmatize(word,pos="v")
    return (str.lstrip())
```

```python
lemmatizer("probably alltime favorite movie story selflessness sacrifice dedication noble cause preachy boring never gets old despite seen 15 times last 25 years paul lukas performance brings tears eyes bette davis one truly sympathetic roles delight kids grandma says like dressedup midgets children makes fun watch mothers slow awakening whats happening world roof believable startling dozen thumbs theyd movie")
```

# Bag Of Word

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(data['review'])
```

```python
y=data["sentiment"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

# ML algo => Logistic Regression

```python
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_pred = log_model.predict(X_test)

print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, log_pred))
print("\nClassification Report:\n", classification_report(y_test, log_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, log_pred))
```

# ML algo=>Random Forest

```python
rf_model = RandomForestClassifier(n_estimators=50, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

print("\nRandom Forest:")
print("Accuracy:", accuracy_score(y_test, rf_pred))
print("\nClassification Report:\n", classification_report(y_test, rf_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, rf_pred))
```
